```python
import pandas as pd
import bottleneck as bn
```

```python
df_pd = pd.read_csv("train.csv")
df_pd
```

```python
print(bn.nanmean(df_pd['Age']))#средний возраст пассажиров
print(bn.nanstd(df_pd['Age']))#его стандартное отклонение
```

```python
def fare_new(value):
    if value['Pclass'] == 1 or value['Pclass'] == 2:
      return value['Fare'] * 1.3
    elif value['Pclass'] == 3:
      return value['Fare'] * 1.1
```

```python
df_pd['Fare_new'] = df_pd.apply(fare_new, axis = 1)
```

# Оптимизация типов pandas

```python
import numpy as np
```

```python
df3 = pd.read_csv("Housing.csv")
df3
```

```python
df3.info() # до смены типов данных
```

```python
# т.к. мы знаем что все числовые данные в таблице не отрицательные и не могут ими быть берем uint
df3['price'] = df3['price'].astype(np.uint32) # максимальное значение 13300000 поэтому берем uint32 (диапазон 0 - 4,2*10^9)
df3['area'] = df3['area'].astype(np.uint16) # максимальное значение 16200 поэтому берем uint16 (диапазон 0 - 65635)
df3['bedrooms'] = df3['bedrooms'].astype(np.uint8) # максимальное значение 6 поэтому берем uint8 (диапазон 0-255)
df3['bathrooms'] = df3['bathrooms'].astype(np.uint8) # максимальное значение 4 поэтому берем uint8
df3['stories'] = df3['stories'].astype(np.uint8) # максимальное значение 4 поэтому берем uint8
df3['parking'] = df3['parking'].astype(np.uint8) # максимальное значение 3 поэтому берем uint8
# для строковых данных используем тип 'category'
df3['mainroad'] = df3['mainroad'].astype("category")
df3['guestroom'] = df3['guestroom'].astype("category")
df3['basement'] = df3['basement'].astype("category")
df3['hotwaterheating'] = df3['hotwaterheating'].astype("category")
df3['airconditioning'] = df3['airconditioning'].astype("category")
df3['prefarea'] = df3['prefarea'].astype("category")
df3['furnishingstatus'] = df3['furnishingstatus'].astype("category")
```

```python
df3.info() # после смены типов данных
```
